1. Imports y preparaciÃ³n

```python
import pandas as pd
import numpy as np

from src.analysis.classifier import NewsClassifier
from src.data_processing.data_validator import DataValidator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

2. Upload dataset of example

```python
data = {
    "text": [
        "The president met with foreign leaders to discuss trade agreements.",
        "The soccer team won the championship after a dramatic final match.",
        "Stock markets crashed today amid fears of recession.",
        "Apple announced the new iPhone model with major camera improvements."
    ],
    "label": ["politics", "sports", "economy", "technology"]
}

df = pd.DataFrame(data)
df
```

3. Validation of data

```python
validator = DataValidator(required_columns=["text", "label"])
validator.validate(df)
```

4. Train-test split

```python
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df["text"],
    df["label"],
    test_size=0.25,
    random_state=42
)
```

5. Train the clasificador

```python
clf = NewsClassifier(language="en")
clf.fit(train_texts.tolist(), train_labels.tolist())
```

6. Evaluation

```python
preds = []
for t in test_texts.tolist():
    preds.append(clf.predict(t)["label"])

print(classification_report(test_labels, preds))
```

7. Test
