1. Imports

```python
import pandas as pd
from src.analysis.topic_modeler import TopicModeler
```

2. Dataset

```python
corpus = [
    "The government passed a controversial new law involving taxes.",
    "New studies show that exercise improves mental health.",
    "Scientists discovered a new exoplanet similar to Earth.",
    "Apple unveils new MacBook with improved performance."
]

df = pd.DataFrame({"text": corpus})
df
```

3. Create topic models

```python
tm = TopicModeler(n_topics=2, language="en")
tm.fit(df["text"].tolist())
```

4. Show the topics

```python
topics = tm.get_topics(n_top_words=8)
topics
```

5. Get distribution of the topics in a document

```python
text = "The new NASA mission will explore Mars next year."
tm.transform_document(text, method="lda")
```
